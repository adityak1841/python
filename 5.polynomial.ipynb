```python
from sklearn.preprocessing import PolynomialFeatures 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression



poly_features=PolynomialFeatures(degree=2)
data=pd.read_csv('Ice_cream selling data.csv')
x=data['A']
y=data['B']


x_poly=poly_features.fit_transform(x.values.reshape(-1,1))#reshape means it converts 1d array into  2d
model=LinearRegression()#to create a linear regression
model.fit(x_poly,y)#to train a model by passing the training data fit function is used 
y_predict=model.predict(x_poly) #it predicts the value of y from trained data 
plt.scatter(x,y,label='Data')
plt.plot(x,y_predict)
plt.show()
#plt.legend()
```
